```python
import pandas as pd
import json

# Caminho do arquivo JSON
caminho_json = "financeiro_ambassador.json"

# Ler o arquivo JSON
with open(caminho_json, "r", encoding="utf-8") as f:
    dados = json.load(f)

# Converter para DataFrame
df = pd.DataFrame(dados)

df.to_json('extrato_ambassador.json', index=False)

# Exibir as primeiras linhas
df.head()
```
